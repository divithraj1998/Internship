```python
def extract_data(n):

    #step2-->connect to a browser
    
    flip=webdriver.Chrome()
    
    #step3 open the required page using driver
    
    flip.get("https://www.flipkart.com/")
    
    #step4-->enter the data 
    
    sunglasses=flip.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
    sunglasses.send_keys("sunglasses")


    search=flip.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button")
    search.click()
    
    
    #step5-->create list
    
    brand_name=[]
    description=[]
    price=[]
    discount=[]
    
    #step6-->scrape the data
    
    for x in range(n):
    
        brand=flip.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
        for i in brand:
            brand_name.append(i.text)
    
        describe=flip.find_elements(By.XPATH,'//a[@class="IRpwTa _2-ICcC" or @class="IRpwTa"]')
        for i in describe:
            description.append(i.text)

        amt=flip.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
        for i in amt:
            price.append(i.text)

        disc=flip.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
        for i in disc:
            discount.append(i.text)
            
        time.sleep(3)
         
        next = flip.find_elements(By.CLASS_NAME,"_1LKTO3")
        
        if len(next)==1:
            next[0].click()
        else :
            next[1].click()
            
        time.sleep(5)
    
    
    df=pd.DataFrame({'BRAND':brand_name,'DESCRIPTION':description,'PRICE':price,'DISCOUNT':discount})
    return(df)
```

```python
sunglasses=extract_data(3)
sunglasses=sunglasses.iloc[0:100,:]
print(sunglasses)
```

```python
def iphone(n):


    flip=webdriver.Chrome()

    
    flip.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")

    rating=[]
    summary=[]
    review=[]
    
    for x in range(n):
    
        rate=flip.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
        for i in rate:
            rating.append(i.text)
    
        summ=flip.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
        for i in summ:
            summary.append(i.text)

        rev=flip.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
        for i in rev:
            review.append(i.text)
            
        time.sleep(3)
        
        next = flip.find_elements(By.CLASS_NAME,"_1LKTO3")
        
        if len(next)==1:
            next[0].click()
        else :
            next[1].click()
    
        #relax and scrape data must use when extracting data from multiple pages
        
        time.sleep(5)
        
    print(len(rating))
    
    print(len(summary))
    
    print(len(review))
    
    #step 7-->creating dataframe
    
    df=pd.DataFrame({'RATING':rating,'SUMMARY':summary,'REVIEW':review})
    return(df)
```

```python
def extract_data(n):

    #step2-->connect to a browser
    
    flip=webdriver.Chrome()
    
    #step3 open the required page using driver
    
    flip.get("https://www.flipkart.com/")
    
    #step4-->enter the data 
    
    sneakers=flip.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
    sneakers.send_keys("sneakers")


    search=flip.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button")
    search.click()
    
    
    #step5-->create list
    
    brand_name=[]
    description=[]
    price=[]
    
    #step6-->scrape the data
    
    for x in range(n):
    
        brand=flip.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
        for i in brand:
            brand_name.append(i.text)
    
        describe=flip.find_elements(By.XPATH,'//a[@class="IRpwTa _2-ICcC" or @class="IRpwTa"]')
        for i in describe:
            description.append(i.text)

        amt=flip.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
        for i in amt:
            price.append(i.text)

        time.sleep(3)
        
        next = flip.find_elements(By.CLASS_NAME,"_1LKTO3")
        
        if len(next)==1:
            next[0].click()
        else :
            next[1].click()
    
        
        time.sleep(5)
        
    #step 7-->creating dataframe
    
    df=pd.DataFrame({'BRAND':brand_name,'DESCRIPTION':description,'PRICE':price})
    return(df)
```

```python
def quotations(n):

    
    azquo=webdriver.Chrome()
    
    
    azquo.get("https://www.azquotes.com/")
    
    
    search=azquo.find_element(By.XPATH,"/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
    search.click()

    
    quote=[]
    author=[]
    type=[]
    
    for x in range(n):
    
        quotes=azquo.find_elements(By.XPATH,'//div[@class="wrap-block"]/p/a[2]')
        for i in quotes:
            quote.append(i.text)
    
        auth=azquo.find_elements(By.XPATH,'//div[@class="author"]')
        for i in auth:
            author.append(i.text)

        type_quo=azquo.find_elements(By.XPATH,'//div[@class="tags"]')
        for i in type_quo:
            type.append(i.text)
        
        if x==9:
            break;
            
        time.sleep(3)
        
        next = azquo.find_elements(By.CLASS_NAME,"next")
        
        next[0].click()

        time.sleep(5)
        
    #step 7-->creating dataframe
    
    df=pd.DataFrame({'QUOTES':quote,'AUTHOR':author,'TYPE':type})
    return(df)
```

```python
quotations(10)
```
