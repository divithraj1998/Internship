```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank=[] #//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]
name=[] #//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]
artist=[] #//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[2]
upload_date=[] #//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]
views=[] #//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]
```

```python
na=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in na:
    name.append(i.text)

for i in range(len(name)):
    rank.append(f"{i+1}")

art=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in art:
    artist.append(i.text)
    
upload=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in upload:
    upload_date.append(i.text)
    
view=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in view:
    views.append(i.text)
```

```python
df=pd.DataFrame({'RANK':rank,'NAME':name,'ARTIST':artist,'UPLOAD DATE':upload_date,'VIEWS':views})
```

```python
imdb=webdriver.Chrome()

imdb.get("https://www.imdb.com/search/title/?title_type=tv_series&sort=num_votes,desc")

```

```python
Name=[]  #//h1[@class="sc-d8941411-0 dxeMrU"]
Year_span=[] #/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a
Genre =[] #/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[1]
Run_time=[] #/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[4]
Ratings=[] #/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[1]
Votes=[] #/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[3]
urls=[]
```

```python
web_links=imdb.find_elements(By.XPATH,'//a[@class="ipc-title-link-wrapper"]')
x=0
for i in web_links:
    
    if x==10:
        break;
    else:
        urls.append(i.get_attribute('href'))
    x=x+1
```

```python
for y in range(len(urls)):
    
    imdb.get(urls[y])
    
    time.sleep(7)
    
    try:
        nam=imdb.find_element(By.XPATH,'//h1[@class="sc-d8941411-0 dxeMrU"]')
        if nam.text==None:
            Name.append("-")
        else:
            Name.append(nam.text)
    except NoSuchElementException as e:
            Name.append("-")
            
    try:
        ys=imdb.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')
        if ys.text==None:
            Year_span.append("-")
        else:
            Year_span.append(ys.text)
    except NoSuchElementException as e:
            Year_span.append("-")
    
    try:
        ge=imdb.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[1]')
        if ge.text==None:
            Genre.append("-")
        else:
            Genre.append(ge.text)
    except NoSuchElementException as e:
            Genre.append("-")
    
    try:
        rt=imdb.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[4]')
        if rt.text==None:
            Run_time.append("-")
        else:
            Run_time.append(rt.text)
    except NoSuchElementException as e:
            Run_time.append("-")
            
    try:
        ra=imdb.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')
        if ra.text==None:
            Ratings.append("-")
        else:
            Ratings.append(ra.text)
    except NoSuchElementException as e:
            Ratings.append("-")
            
    try:
        vo=imdb.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[1]/a/span/div/div[2]/div[3]')
        if vo.text==None:
            Votes.append("-")
        else:
            Votes.append(vo.text)
    except NoSuchElementException as e:
            Votes.append("-")
```

```python
df5=pd.DataFrame({'NAME':Name,'YEAR':Year_span,'GENRE':Genre,'RUN TIME':Run_time,'RATINGS':Ratings,'VOTES':Votes})
df5
```

```python
uci=webdriver.Chrome()

uci.maximize_window()

uci.get("https://archive.ics.uci.edu/ You")

```

```python
dataset=[] #//h2[@class="truncate text-primary"]
data_type=[] # //div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[1]/span
task=[] # //div[@class="rounded-box bg-base-100"]/div/div[2]/p
attribute=[] # //div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[2]/span
instance=[] # //div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[3]/span
no_att=[] # //div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[4]/span
year=[] # //div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]
```

```python
datasets=uci.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div/div/div/a[2]')
datasets.click()
```

```python
for x in range(100):
    
    expand=uci.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[1]/div/label[2]')
    expand.click()
    
    dat=uci.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]')
    for i in dat:
        dataset.append(i.text)
    
    typ=uci.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[1]/span')
    for i in typ:
        data_type.append(i.text)

    tas=uci.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/p')
    for i in tas:
        task.append(i.text)
    
    att=uci.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
    for i in att:
        attribute.append(i.text)
    
    ins=uci.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[3]/span')
    for i in ins:
        instance.append(i.text)
    
    no=uci.find_elements(By.XPATH,'//div[@class="rounded-box bg-base-100"]/div/div[2]/div/div[4]/span')
    for i in no:
        no_att.append(i.text)
    
    ye=uci.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')
    for i in ye:
        year.append(i.text)
        
    for m in range(20):
        uci.execute_script("window.scrollBy(0,500)")
        
    try:
        next=uci.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
        next.click()
    except ElementClickInterceptedException as e:
        break;
        
    time.sleep(7)
    
```

```python
df6=pd.DataFrame({'DATA SET':dataset,'DATA TYPE':data_type,'TASK':task,'TYPE OF ATTRIBUTE':attribute,'NO OF INSTANCES':instance,'NO OF ATTRIBUTE':no_att,'YEAR':year})
df6.head(100)
```
